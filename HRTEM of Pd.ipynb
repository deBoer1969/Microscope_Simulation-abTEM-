```python
%config InlineBackend.rc = {"figure.dpi": 72, "figure.figsize": (6.0, 4.0)}
%matplotlib inline

import abtem
import ase
import matplotlib.pyplot as plt
import numpy as np
from ase.io import read
from ase.visualize import view
```

```python
atoms = ase.io.read(r"C:\Users\Joshu\Desktop\For python\Simulated_Moleculare_Models\Pd_1000_NP.cfg")
```

```python
view(atoms)
```

```python
abtem.config.set({"device": "cpu", "fft": "fftw"})
```

```python
frozen_phonons = abtem.FrozenPhonons(atoms, 16, sigmas=0.1)
```

```python
potential = abtem.Potential(
    frozen_phonons,
    sampling=0.05,
    projection="infinite",
    slice_thickness=0.5,
)
```

```python
wave = abtem.PlaneWave(energy=100e3)
```

```python
exit_wave = wave.multislice(potential)
```

```python
exit_wave.compute()
```

```python
exit_wave.show(cmap='gray')
```
